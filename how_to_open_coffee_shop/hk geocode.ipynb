```python
import requests
from bs4 import BeautifulSoup
search_phrase = 'cyberport'
url = r'https://www.als.ogcio.gov.hk/lookup?q=' + search_phrase
response = requests.get(url)
# soup = BeautifulSoup(response.text,'lxml')
soup = BeautifulSoup(response.text)
```

```python
# construct element tree
for tag in soup.findChildren():
    print(tag.name)
    print(tag.contents)
    
    print("-"*80)
```
